```python
import plotly.express as px
df = px.data.iris()
fig = px.scatter(df, x = "sepal_width", y = "sepal_length", color = "species", marginal_y = "violin", marginal_x = "box", trendline="ols", 
                 template="simple_white")
fig.show()
```

```python
import plotly.express as px
df = px.data.gapminder()
fig = px.scatter(
    df, x="gdpPercap", y="lifeExp", size="pop", color="continent", hover_name="country", facet_col="continent",animation_frame="year",
    animation_group="country", log_x=True, size_max=45, range_x=[100,100000], range_y=[25,90])
fig.show()
```
