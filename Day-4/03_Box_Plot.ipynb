```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

titanic = sns.load_dataset("titanic")

sns.boxplot(x="survived",y="age", data=titanic , 
               showmeans=True , 
               meanprops={"marker":"+","markersize":"12","markeredgecolor":"red"})

#show Labels
plt.xlabel("How many survived")
plt.ylabel("What was the age")
plt.title("The BarPlot")



plt.show()
```

# facet wrap and facet grid ?

```python
import seaborn as sns
import matplotlib.pyplot as plt

tip= sns.load_dataset("tips")

p1=sns.FacetGrid(tip, col="sex",hue="day")
plt.figure(figsize=(12,10))

p1.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```
