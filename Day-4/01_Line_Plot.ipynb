# Import Libraries


```python
#importing libraries

import seaborn as sns
import matplotlib.pyplot as plt 

#loading DataSet

phool = sns.load_dataset("iris")

phool

#Draw a Line Plot

sns.lineplot(data=phool, x="sepal_length" , y="sepal_width")
plt.show()
```

# Adding Titles

```python
#importing libraries

import seaborn as sns
import matplotlib.pyplot as plt 

#loading DataSet

phool = sns.load_dataset("iris")

phool

#Draw a Line Plot

sns.lineplot(data=phool, x="sepal_length" , y="sepal_width")
plt.title("Iris Plot")

plt.show()
```

# Adding Limits

```python
#importing libraries

import seaborn as sns
import matplotlib.pyplot as plt 

#loading DataSet

phool = sns.load_dataset("iris")

phool

#Draw a Line Plot

sns.lineplot(data=phool, x="sepal_length" , y="sepal_width")
plt.title("Iris Plot")

plt.xlim(2)
plt.ylim(1)

plt.show()
```

## Set Styles


```python
#importing libraries

import seaborn as sns
import matplotlib.pyplot as plt 



#loading DataSet

phool = sns.load_dataset("iris")

phool

#Draw a Line Plot

sns.lineplot(data=phool, x="sepal_length" , y="sepal_width")
plt.title("Iris Plot")
sns.set_style(style=None,rc=None)
sns.set_style("dark")

sns.set_style("ticks")

plt.show()
```

## Size of the Figure

```python
#importing libraries

import seaborn as sns
import matplotlib.pyplot as plt 

#loading DataSet

phool = sns.load_dataset("iris")

#Draw a Line Plot
plt.figure(figsize=(10,6))

sns.lineplot(data=phool, x="sepal_length" , y="sepal_width")
plt.title("Iris Plot")



plt.show()
```
